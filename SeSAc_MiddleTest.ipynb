### 1번 문제 : 데이터분석


**sales_data.csv 파일을 업로드하거나 아래 코드를 한번 실행하여 생성해주세요.**

</br>

주어진 sales_data.csv 파일에는 월별 매출 데이터가 포함되어 있습니다.

각 열에는 Month(월), Product(제품명), Sales(판매액) 정보가 있습니다.

</br>
csv 파일을 읽고 아래 3가지 값을 print()하세요.


1. 제품별 매출 합계 구하기


2. 매출이 가장 높은 제품을 구하기


3. 월별 평균 매출을 계산

```python
import pandas as pd

data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']*3,
    'Product': ['A']*12 + ['B']*12 + ['C']*12,
    'Sales': [150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650,
              80, 120, 160, 200, 240, 280, 320, 360, 400, 440, 480, 520]
}

df = pd.DataFrame(data)

file_path = 'sales_data.csv'
df.to_csv(file_path, index=False)

file_path
```

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv("sales_data.csv")

# 제품별 매출 합계 구학
productSales = df.groupby("Product")["Sales"].sum()
print("\n제품별 매출 합계는: ")
print(productSales)

# 매출이 가장 높은 제품 구하기
bestProduct = df.loc[df["Sales"].idxmax()]["Product"]
print("\n매출이 가장 높은 제품은:")
print(bestProduct)

# 월별 제품별 평균 매출 구하기
monthlySales = df.groupby(["Month", "Product"])["Sales"].mean().unstack()
print("\n월별 제품 평균 매출은: ")
print(monthlySales)
```

### 2번 문제 : 머신러닝


**customer_churn.csv 파일을 업로드하거나 아래 코드를 한번 실행하여 생성해주세요.**

</br>


주어진 customer_churn.csv 데이터에는 고객의 다양한 특성과 이탈 여부가 포함되어 있습니다.

이 데이터를 바탕으로 로지스틱 회귀 모델을 학습시키고, 테스트 데이터에 대한 예측을 수행하는 코드를 작성하세요.

예측 결과 : 고객의 이탈 여부


</br>
csv 파일을 읽고 모델 학습 후 모델의 accuracy 평가 값을 print()하세요. (단, 데이터셋을 훈련과 테스트 데이터로 분리 시 random_state=1로 넣어야 합니다.)


```python
import pandas as pd

data = {
    'CustomerID': range(1, 101),
    'Age': [i % 50 + 20 for i in range(100)],  # 20에서 70 사이의 나이
    'MonthlyCharges': [50 + i % 50 for i in range(100)],  # 월별 구매 금액
    'Tenure': [1 + i % 72 for i in range(100)],  # 1에서 72개월 사이의 서비스 이용 기간
    'Churn': [1 if (i % 2 == 0 and i % 5 == 0) else 0 for i in range(100)]  # 고객의 이탈 여부
}

df_churn_balanced = pd.DataFrame(data)

file_path_churn_balanced = 'customer_churn.csv'
df_churn_balanced.to_csv(file_path_churn_balanced, index=False)

file_path_churn_balanced
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# CSV 파일 읽기
df = pd.read_csv("customer_churn.csv")

# 필요한 열 선택 및 전처리
x = df[["Age", "MonthlyCharges", "Tenure"]]
y = df["Churn"]

# 훈련용 데이터와 테스트용 데이터로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

# 모델 학습
mymodel = LogisticRegression()
mymodel.fit(x_train, y_train)

# 테스트 데이터의 결과 예측
y_pred = mymodel.predict(x_test)

# 정확도 평가 출력
accuracy = accuracy_score(y_test, y_pred)
print(f"모델의 정확도: {accuracy:.2f}")
```

```python
import cv2
import matplotlib.pyplot as plt

# 주어진 원본 img.jpg를 찾을 수 없으므로, 임의의 이미지를 내려받아 사용하기로 합니다
image_path = "coolAnya.jpg"
image = cv2.imread(image_path)

# Haar Cascade 파일 경로
cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'

# Haar Cascade 분류기 로드
face_cascade = cv2.CascadeClassifier(cascade_path)

# 얼굴 감지
faces = face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# 감지된 얼굴에 사각형 테두리 그리기
for (x, y, w, h) in faces:
  cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# 이미지 변환 및 저장
image_r = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
output_path = "coolAnyaGoodFace.jpg"
cv2.imwrite(output_path, image)

# 결과 이미지 표시
plt.imshow(image_r)
plt.axis("off")
plt.show()
```
